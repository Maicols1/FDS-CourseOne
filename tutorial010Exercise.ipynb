## Consider a new type of noise. 

Next, we will want to add a different form of noise to the simulated planes, to do so, we will reuse the planes simulated above.

The noise that we added to the simulated airplanes in tutorial 010 and above is *independent* across airplanes and time points. This type of noise is technically called *IID* or (Independent and Identically Distributed noise)[https://en.wikipedia.org/wiki/Independent_and_identically_distributed_random_variables].

Yet, if we think a more realistic scenario, several planes (espacially the ones departing and landing close to each others) might have noise coming from the environment (the city) that is not IID but that is common to all planes. For example, a leafblower in the neighborhood working for a couple of hours will affect a few planes and add noise that is common to the airplanes. Let's call this noise *shared noise*.

Your task is to simulate ten (airplanes just like above) that whose measurements are affected by both IID noise and shared noise.

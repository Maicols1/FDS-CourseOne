### multiple variables in columns

Specity whatever filename you want:

```python
fname = "testDatFile.csv"
```

Make some data:

```python
n = 100 # number of observations
nv = 3  # number of variables
# Import - need to specify col names manually until we level up the code
myData = np.random.randn(n, nv)
myData[:,0] = 1 + 1.5*myData[:,0] # give one var a higher mean and sd
```

Look at what we're going to put in the file:

```python
myData.shape
```

```python
plt.plot(myData)
```

Now make the numpy array into a pandas data frame:

```python
myDataFr = pd.DataFrame(myData, columns=["VarA", "VarB", "VarC"])
```

Save the data frame to a .csv (comma separated values) file:

```python
myDataFr.to_csv(fname, index=False)
```
